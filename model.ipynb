```python
def clean_column_name(col):
    # 1. Remove special characters at the start (like @, #, $, etc.)
    col = re.sub(r'^[^a-zA-Z0-9]+', '', col)
    
    # 2. Replace spaces and special characters in between with underscore
    col = re.sub(r'[^a-zA-Z0-9]+', '_', col)
    
    # 3. Convert to lowercase
    col = col.lower()
    
    # 4. Remove trailing underscores (if any)
    col = col.strip('_')
    
    return col
df.columns = [clean_column_name(col) for col in df.columns]

# Verify result
print("âœ… Cleaned Column Names:\n", list(df.columns))
```

```python
unnamed_cols = [col for col in df.columns if 'unnamed' in col.lower()]
print("Unnamed columns:", unnamed_cols)

# Peek at their data
df[unnamed_cols].head(10)
```

```python
# Identify all 'Unnamed' columns
unnamed_cols = [col for col in df.columns if 'unnamed' in col.lower()]

# Summary report for unnamed columns
print("ðŸ§¾ Unnamed Column Summary:\n")

summary = []
for col in unnamed_cols:
    total = len(df)
    missing = df[col].isnull().sum()
    unique_vals = df[col].nunique(dropna=True)
    
    # Most frequent value
    mode_val = df[col].mode().iloc[0] if unique_vals > 0 else None
    
    summary.append({
        "Column": col,
        "Missing_Count": missing,
        "Missing_%": round((missing / total) * 100, 2),
        "Unique_Values": unique_vals,
        "Most_Common_Value": mode_val
    })

summary_df = pd.DataFrame(summary)
print(summary_df)
```

```python
# Identify low-information columns
drop_cols = []

for col in [c for c in df.columns if 'unnamed' in c.lower()]:
    if df[col].nunique(dropna=True) == 1:
        drop_cols.append(col)
    elif df[col].nunique(dropna=True) < 5 and df[col].mode()[0].lower() in ["#", "not assigned", "na", "none"]:
        drop_cols.append(col)

# Drop those columns
df.drop(columns=drop_cols, inplace=True)
print("ðŸ§¹ Dropped columns:", drop_cols)
```

```python
# Loop through all unnamed columns and display top 10 frequent values
for col in unnamed_cols:
    print(f"\nðŸ“Š Frequency distribution for column: {col}")
    print(df[col].value_counts(dropna=False).head(10))
    print("-" * 60)
```

```python
# 1ï¸âƒ£ Detect columns with at least one "Not assigned" value
cols_to_convert = [col for col in df.columns if df[col].astype(str).str.lower().str.contains('not assigned').any()]

print("Columns to convert:", cols_to_convert)

# 2ï¸âƒ£ Fill NaN with "Not assigned" to handle missing values
df[cols_to_convert] = df[cols_to_convert].fillna("Not assigned")

# 3ï¸âƒ£ Convert to binary (0 = "Not assigned", 1 = any actual value)
for col in cols_to_convert:
    df[col] = df[col].apply(lambda x: 0 if str(x).strip().lower() == "not assigned" else 1)

# 4ï¸âƒ£ Verify the conversion
print(df[cols_to_convert].head(10))
```

```python
import pandas as pd

# -----------------------------
# PARAMETERS
# -----------------------------
customer_col = 'customer_id'      # Customer identifier
sales_col = 'sales_volume_kg'     # Metric to track activity
period_col = 'fiscal_period'      # Time column
num_periods = 3                   # Last N periods to check for churn

# -----------------------------
# STEP 1: Identify last N fiscal periods
# -----------------------------
last_periods = sorted(df[period_col].unique())[-num_periods:]
print(f"Using last {num_periods} fiscal periods for churn: {last_periods}")

# -----------------------------
# STEP 2: Filter dataset for last N periods
# -----------------------------
recent_df = df[df[period_col].isin(last_periods)]

# -----------------------------
# STEP 3: Aggregate purchases per customer
# -----------------------------
customer_agg = recent_df.groupby(customer_col)[sales_col].sum().reset_index()

# -----------------------------
# STEP 4: Define churn target
# 1 = no purchase in last N periods, 0 = purchased at least once
# -----------------------------
customer_agg['churn'] = customer_agg[sales_col].apply(lambda x: 1 if x == 0 else 0)

# -----------------------------
# STEP 5: Optional - drop sales column, keep only target
# -----------------------------
churn_target = customer_agg[[customer_col, 'churn']]

# -----------------------------
# STEP 6: Preview
# -----------------------------
print(churn_target.head())
print(f"Churn distribution:\n{churn_target['churn'].value_counts(normalize=True)}")
```
