```python
def clean_column_name(col):
    # 1. Remove special characters at the start (like @, #, $, etc.)
    col = re.sub(r'^[^a-zA-Z0-9]+', '', col)
    
    # 2. Replace spaces and special characters in between with underscore
    col = re.sub(r'[^a-zA-Z0-9]+', '_', col)
    
    # 3. Convert to lowercase
    col = col.lower()
    
    # 4. Remove trailing underscores (if any)
    col = col.strip('_')
    
    return col
df.columns = [clean_column_name(col) for col in df.columns]

# Verify result
print("✅ Cleaned Column Names:\n", list(df.columns))
```

```python
# Convert fiscal period to datetime (assuming format like 'April 2023')
df['fiscal_period'] = pd.to_datetime(df['fiscal_period'], format='%B %Y')

# Make sure sales columns are numeric
df['Actual_Sales_Volume_in_kgs'] = pd.to_numeric(df['Actual_Sales_Volume_in_kgs'], errors='coerce')
df['actual_net_sales_in_eur'] = pd.to_numeric(df['actual_net_sales_in_eur'], errors='coerce')

# Fill NaNs with 0
df[['Actual_Sales_Volume_in_kgs', 'actual_net_sales_in_eur']] = df[['Actual_Sales_Volume_in_kgs', 'actual_net_sales_in_eur']].fillna(0)

# Sort for time-based calculations
df = df.sort_values(by=['customer', 'fiscal_period'])
```

```python
last_purchase = (
    df.groupby('customer')['fiscal_period']
    .max()
    .reset_index()
    .rename(columns={'fiscal_period': 'last_purchase_date'})
)

df = df.merge(last_purchase, on='customer', how='left')
```

```python
analysis_month = pd.to_datetime("april 2025")
start_period = analysis_month - pd.DateOffset(months=12)

# Filter for the last 12 months
recent_df = df[(df['fiscal_period'] > start_period) & (df['fiscal_period'] <= analysis_month)]

# Count distinct active months per customer
active_months = (
    recent_df.groupby('customer')['fiscal_period']
    .nunique()
    .reset_index()
    .rename(columns={'fiscal_period': 'active_months_last_12m'})
)
```

```python
def categorize_frequency(months):
    if pd.isna(months):  # No history
        return "Monthly"
    elif months >= 10:
        return "Monthly"
    elif 4 <= months <= 9:
        return "Quarterly"
    elif 2 <= months <= 3:
        return "Half-Yearly"
    elif months <= 1:
        return "Yearly"

active_months['Customer_Frequency'] = active_months['active_months_last_12m'].apply(categorize_frequency)
```

```python
df = df.merge(active_months, on='customer', how='left')
```

```python
monthly_sales = (
    df.groupby(['customer', 'fiscal_period'])
    .agg({'Actual_Sales_Volume_in_kgs': 'sum'})
    .reset_index()
)
```

```python
def get_period_range(frequency, analysis_month):
    if frequency == "Monthly":
        current_start = analysis_month.replace(day=1)
        current_end = analysis_month
        prev_start = current_start - pd.DateOffset(months=1)
        prev_end = current_start - pd.DateOffset(days=1)
    elif frequency == "Quarterly":
        current_start = analysis_month - pd.DateOffset(months=2)
        current_end = analysis_month
        prev_start = current_start - pd.DateOffset(months=3)
        prev_end = current_start - pd.DateOffset(days=1)
    elif frequency == "Half-Yearly":
        current_start = analysis_month - pd.DateOffset(months=5)
        current_end = analysis_month
        prev_start = current_start - pd.DateOffset(months=6)
        prev_end = current_start - pd.DateOffset(days=1)
    else:  # Yearly
        current_start = analysis_month - pd.DateOffset(months=11)
        current_end = analysis_month
        prev_start = current_start - pd.DateOffset(months=12)
        prev_end = current_start - pd.DateOffset(days=1)
    return current_start, current_end, prev_start, prev_end
```

```python
def compute_churn(customer, freq):
    current_start, current_end, prev_start, prev_end = get_period_range(freq, analysis_month)
    
    cust_data = monthly_sales[monthly_sales['customer'] == customer]
    
    # Current and previous sales
    current_sales = cust_data[
        (cust_data['fiscal_period'] >= current_start) & (cust_data['fiscal_period'] <= current_end)
    ]['Actual_Sales_Volume_in_kgs'].sum()
    
    prev_sales = cust_data[
        (cust_data['fiscal_period'] >= prev_start) & (cust_data['fiscal_period'] <= prev_end)
    ]['Actual_Sales_Volume_in_kgs'].sum()
    
    # Churn conditions
    no_orders_12m = current_sales == 0
    drop_20 = (prev_sales > 0) and ((current_sales / prev_sales) < 0.8)
    
    churn_flag = int(no_orders_12m or drop_20)
    
    return pd.Series({
        'current_sales': current_sales,
        'prev_sales': prev_sales,
        'is_churn': churn_flag
    })

# Apply churn logic to each customer
results = []
for _, row in active_months.iterrows():
    cid = row['customer']
    freq = row['Customer_Frequency']
    result = compute_churn(cid, freq)
    result['customer'] = cid
    result['Customer_Frequency'] = freq
    results.append(result)

churn_df = pd.DataFrame(results)
```

```python
final_df = df.merge(churn_df, on=['customer', 'Customer_Frequency'], how='left')

# Keep only unique customer-level info
final_customer_view = final_df.drop_duplicates(subset=['customer'])[
    ['customer', 'last_purchase_date', 'active_months_last_12m',
     'Customer_Frequency', 'current_sales', 'prev_sales', 'is_churn']
]

final_customer_view.head()
```
